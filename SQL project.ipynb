```python
#importing libraries
import pandas as pd
import sqlite3
import statistics
```

```python
#fetching csv file with pandas data frame
df = pd.read_csv("gobankingrates.csv")

```

```python
df
```

```python
#connecting to sqlite3
conn = sqlite3.connect("database.db")
#making df into sql table
df.to_sql("city_table", conn, if_exists="replace")
```

```python
#readjusting df variable to return all columns from city_table
df = pd.read_sql_query("SELECT * FROM city_table", conn)
```

```python
#checking to see if it worked
df
```

```python
#returning all data types
print(df.dtypes)
```

```python
#making city_homeowner column into a list
s = df['city_homeowner'].tolist()
```

```python
s
```

```python
#extracting all the number values from the list
values = []

for v in s:
    
    num = v[-10:]
    
    values.append(num)
```

```python
values
```

```python
#removing "$" and "," so that i can convert each value into a float
home_owner = []

for v in values:
    
    
    s = v.replace("$", "").replace(",", "")
        
    float_value = float(s)
    
    home_owner.append(float_value)
    
    
    
```

```python
home_owner
```

```python
#now that every value is a float i can find the average
average = statistics.mean(home_owner)
```

```python
average
```

```python
#returning all the values that are below the average and adding "$" and "," back into each value
new_list_h = [format("${:,.2f}".format(value)) for value in home_owner if value < average]
```

```python
#converting each value into string
ch = []

for v in new_list_h:
    
    s_value = str(v)
    
    ch.append(s_value)
```

```python
ch
```

```python
#adding proper format for each value
for i in range(len(ch)):
    ch[i] = "Income needed if you’re a homeowner: " + ch[i]
```

```python
ch
```

```python
#creating a query that returns each value of my ch list
query_1 = "SELECT * FROM city_table WHERE city_homeowner IN ({})".format(', '.join(['?' for x in ch]))

d = pd.read_sql_query(query_1, conn, params=ch)
```

```python
d
```

```python
#converting 'city_renter' into a list
string = df['city_renter'].tolist()
```

```python
#extracting all the number values from the list
r_values = []

for v in string:
    
    num = v[-10:]
    
    r_values.append(num)
```

```python
r_values
```

```python
#removing "$" and "," so that i can convert each value into a float
renter = []

for v in r_values:
    
    
    s = v.replace("$", "").replace(",", "")
        
    float_value = float(s)
    
    renter.append(float_value)
```

```python
renter
```

```python
#now that every value is a float i can find the average
average_2 = statistics.mean(renter)
```

```python
average_2
```

```python
#returning all values below average and re inserting "$" and "," to each value 
new_list_r = [format("${:,.2f}".format(num)) for num in renter if num < average_2]
```

```python
#converting each value into a string
rh = []

for r in new_list_r:
    
    r_value = str(r)
    
    rh.append(r_value)
```

```python
rh
```

```python
#using for loop to add proper format to each value 
for i in range(len(rh)):
    rh[i] = "Income needed if you’re a renter: " + rh[i]
```

```python
rh
```

```python
#creating a query that returns each value from my rh list
query_2 = "SELECT * FROM city_table WHERE city_renter IN ({})".format(", ".join(["?" for x in rh]))

renter_d = pd.read_sql_query(query_2, conn, params=rh)
```

```python
renter_d
```

```python
d
```

```python
#using pandas to inner merge my 2 queries
result = pd.merge(renter_d, d, on=None, how='inner')
```
